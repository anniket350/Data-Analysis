```python
# Importing all libraries required in this notebook
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
import warnings
```

```python
df=pd.read_csv('SampleSuperstore.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.groupby("Region")["Sales"].sum().plot.bar()
```

```python
df.groupby("Region")["Profit"].sum().plot.bar()
```

```python
df.groupby("Region")["Sales"].sum().plot.pie(autopct="%0.1f%%")
```

```python
df.groupby("Region")["Profit"].sum().plot.pie(autopct="%0.1f%%")
```

```python
df.groupby("State")["Profit"].sum().plot.bar()
```

```python
plt.figure(figsize= (15,8))
df.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize= (15,8))
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(), annot=True)
```

```python
# Convert 'Sub-Category' column to categorical
df['Sub-Category'] = df['Sub-Category'].astype('category')
```

```python
# Plot countplot
plt.figure(figsize=(8, 7))
sns.countplot(data=df, x='Sub-Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 7))
sns.lineplot(x='Discount', y='Quantity', data=df)
plt.show()
```

```python
plt.figure(figsize=(8, 7))
sns.lineplot(x='Discount', y='Profit', data=df)
plt.show()
```

```python
plt.figure(figsize=(8, 7))
sns.lineplot(x='Ship Mode', y='Profit', data=df)
plt.show()
```

```python
plt.figure(figsize=(15,8))
df.groupby('State')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize= (10,16))
df.groupby('Ship Mode')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Sale')
plt.show()
```
